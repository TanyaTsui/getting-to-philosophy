# Generate 1000 random wikipedia links
run only once

```python
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
import os
import re
```

```python
wikis = []
for i in range(1000): 
    url = 'https://en.wikipedia.org/wiki/Special:Random'
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    title = soup.find('title').string.replace(' - Wikipedia', '').replace(' ', '_')
    title = 'https://en.wikipedia.org/wiki/' + title
    wikis.append(title)
```

```python
len(wikis)
```

```python
import json
with open("randomWikiLinks.txt", "w") as fp:
    json.dump(wikis, fp)
```

```python
with open("randomWikiLinks.txt", "r") as fp:
    wikis = json.load(fp)
```
